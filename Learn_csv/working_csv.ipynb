```python
# Using requests (For Complex Cases)

import pandas as pd  
import requests  
from io import StringIO  

# Fetch data from the URL
url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
response = requests.get(url)

# Convert response content to a DataFrame
csv_data = StringIO(response.text)
df = pd.read_csv(csv_data)
df.head()
```

9.To change any column value by making a function easily.

```python
def rename(education_level):
    if education_level=="Graduate":
        return "GRD_new"
    else:
        return education_level
```

```python
rename('Graduate')
```

```python
df8=pd.read_csv('aug_train.csv',converters={'education_level':rename})
df8
```
