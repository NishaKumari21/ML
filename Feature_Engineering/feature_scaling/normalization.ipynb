```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('wine_data.csv',usecols=[1,2,3],header=None)
df.columns=['Class label', 'Alcohol', 'Malic acid']
df.head()
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
sns.kdeplot(df['Malic acid'])
```

```python
sns.kdeplot(df['Class label'])
```

```python
color={1:'red',2:'green',3:'yellow'}
sns.scatterplot(x=df['Alcohol'],y=df['Malic acid'],hue=df['Class label'],palette=color)
```

```python
color_dict={1:'red',3:'green',2:'blue'}
sns.scatterplot(df['Alcohol'],df['Malic acid'],hue=df['Class label'],palette=color_dict)
```

```python
df.head()
```

```python
x=df.drop(columns=['Class label'])
y=df['Class label']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_train.shape,x_test.shape
```

```python
# standardization 
from sklearn.preprocessing import MinMaxScaler
scalar=MinMaxScaler()
```

```python
scalar.fit(x_train)
x_train_scaled=scalar.transform(x_train)
x_test_scaled=scalar.transform(x_test)
```

```python
x_train
# it is in dataframe
```

```python
x_train_scaled
# it is in numpy
```

```python
# convert the x_train_scaled and x_test_scaled to dataframe
x_train_scaled = pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled = pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
x_train_scaled
# now converted to dataframe
```

```python
np.round(x_train.describe())
# see min and max
```

```python
np.round(x_train_scaled.describe())
# here we can see min:0 and max:1 for noth alcohol and malic acid
```

```python
# construct scatterplot to see before and after normaliation
fig,axes=plt.subplots(1,2,figsize=(15,5))
sns.scatterplot(x=x_train['Alcohol'],y=x_train['Malic acid'],ax=axes[0])
axes[0].set_title(['before'])
# see the min and max is different

sns.scatterplot(x=x_train_scaled['Alcohol'],y=x_train_scaled['Malic acid'],ax=axes[1],color='red')
axes[1].set_title(['After'])
# see the min is 0 and max is 1
```

```python
fig,axes=plt.subplots(1,2,figsize=(15,5))
sns.scatterplot(x=x_train['Alcohol'],y=x_train['Malic acid'],ax=axes[0],c=y_train,cmap='viridis')
axes[0].set_title(['before'])
# see the min and max is different

sns.scatterplot(x=x_train_scaled['Alcohol'],y=x_train_scaled['Malic acid'],ax=axes[1],c=y_train,cmap='viridis')
axes[1].set_title(['After'])

#  c=y_train  Assigns different colors based on class labels.
#  cmap='viridis'  Defines a color gradient for classes.
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))
ax1.set_title(['before'])
sns.kdeplot(x_train['Alcohol'],ax=ax1)
sns.kdeplot(x_train['Malic acid'],ax=ax1)
ax2.set_title(['After'])
sns.kdeplot(x_train_scaled['Alcohol'],ax=ax2)
sns.kdeplot(x_train_scaled['Malic acid'],ax=ax2)
# they lie on different scale before but now they lie on same cale
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(15,5))
ax1.set_title("before Alcohol")
sns.kdeplot(x_train['Alcohol'],ax=ax1)
ax2.set_title("after Alcohol")
sns.kdeplot(x_train_scaled['Alcohol'],ax=ax2)
# there is no change except the scaling part
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(15,5))
ax1.set_title("beforeeeeeeee")
sns.kdeplot(x_train['Malic acid'],ax=ax1)

ax2.set_title("after Malic acid")
sns.kdeplot(x_train_scaled['Malic acid'],ax=ax2)
```
