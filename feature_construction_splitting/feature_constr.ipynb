```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('train.csv',usecols=['Survived','SibSp','Age','Parch','Pclass'])
df.sample(5)
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
x=df.drop(columns=['Survived'])
y=df['Survived']
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
np.mean(cross_val_score(LogisticRegression(),x,y,scoring='accuracy',cv=20))
```

```python
x['Family_size']=x['Parch']+x['SibSp']+1
```

```python
x
```

```python
def family_f(num):
    if num==1:
        return 0
    elif num>1 & num>=4:
        # 'small_fam'
        return 1
    else:
        # 'large_fam'
        return 2
```

```python
family_f(6)
```

```python
x['Family_type']=x['Family_size'].apply(family_f)
```

```python
x
```

```python
x.drop(columns=['SibSp','Parch','Family_size'],inplace=True)
```

```python
x
```

```python
np.mean(cross_val_score(LogisticRegression(),x,y,scoring='accuracy',cv=20))
# accuracy increases when fature construction is done
```
