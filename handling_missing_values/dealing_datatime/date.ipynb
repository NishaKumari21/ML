```python
import numpy as np 
import pandas as pd 
```

```python
date=pd.read_csv('orders.csv')
date.head()
```

```python
date.info()
```

```python
# first convert the date column from Dtype:object to DateTime
date['date']=pd.to_datetime(date['date'])
date.info()
```

```python
# to extract year
date['date_year']=date['date'].dt.year
date.head()
# new column date_year created
```

```python
# to extract month
date['date_month']=date['date'].dt.month
```

```python
# to extract day
date['date_day']=date['date'].dt.day
```

```python
date.head()
```

```python
# to extract daysofweek
date['date_doweeks']=date['date'].dt.weekday
```

```python
# day of week - name

date['date_dow_name'] = date['date'].dt.day_name()
```

```python
# to extract week of the year
date['date_week_year']=date['date'].dt.year
```

```python
date['date_dow_name'].unique
```

```python
# to check it is weekend or not?
date['date_weekend']=np.where(date['date_dow_name'].isin(['Saturday','Sunday']),1,0)
```

```python
date.head()
```

```python
# to extract quaeter of the year
date['date_quarter']=date['date'].dt.quarter
```
