```python
import numpy as np 
import pandas as pd 
```

```python
df=pd.read_csv(r'C:\Users\HP\Desktop\ML_100\agglomerative_clustering\hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv',usecols=['Annual Income (k$)','Spending Score (1-100)'])
```

```python
df.sample(3)
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.title("dendogram")
den=shc.dendrogram(shc.linkage(df,method='ward'))
```

```python
# here k=5 by seeing the dendrogram
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=5,linkage='ward')
```

```python
labels_=cluster.fit_predict(df)
```

```python
labels_
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(df['Annual Income (k$)'],df['Spending Score (1-100)'] ,c=cluster.labels_, cmap='rainbow')
```
