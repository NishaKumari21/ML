```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
df=load_iris()
x=df.data
y=df.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

## using DecisionTreeClassifier as a base_estimator

```python
from sklearn.tree import DecisionTreeClassifier
base_estimator=DecisionTreeClassifier()
```

```python
from sklearn.ensemble import BaggingClassifier
bag_model=BaggingClassifier(estimator=base_estimator,n_estimators=10,random_state=42)
```

```python
bag_model.fit(x_train,y_train)
```

```python
y_pred=bag_model.predict(x_test)
```

## accuracy score

```python
from sklearn.metrics import accuracy_score
accuracy_score=accuracy_score(y_test,y_pred)
accuracy_score
```

## classification_report

```python
from sklearn.metrics import classification_report
classification_report(y_test,y_pred,target_names=df.target_names)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## showing confusion matrix using heatmap

```python
sns.heatmap(cm,annot=True,cmap='Purples',fmt='d')
```

* fmt : it is used for the number formatting inside the cell of confusion matrix
* 'd'=for decimal format
* '0.2f'=floating format
