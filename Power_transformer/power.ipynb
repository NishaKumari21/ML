### Purpose of rÂ² Score (Coefficient of Determination):
* It tells how well the independent variables explain the variance in the dependent variable.

* The value ranges from -âˆž to 1:
* 1 â†’ Perfect fit (predictions are exactly correct)
* 0 â†’ Model is no better than a random guess
* Negative â†’ Model performs worse than a simple mean-based prediction

```python
from sklearn.preprocessing import PowerTransformer
```

### PowerTransformer koi v distribution ko gussian distribution(normal distribution mei convert krne k kaam aata h) mtlb  skewed data k liye use hota h
* ðŸ“Œ Types of Power Transformations in Sklearn
*  Box-Cox Transformation (Works only for positive values)
*  Yeo-Johnson Transformation (Works for both positive & negative values)

* When to Use Power Transformer?
* If data is skewed (e.g., exponential or log-normal distribution).
* When model performance is poor due to non-normality.
*Especially useful for linear models (Regression, PCA, etc.) that assume normally distributed data.

 * When NOT to Convert to Normal Distribution?
 * If using Tree-based models (Decision Tree, Random Forest, XGBoost) â†’ They donâ€™t require normally distributed data.
 * If the original distribution has a natural meaning (e.g., Binary classification, Categorical data).

## Box cox tranform

```python
# Applying Box-Cox Transform

pt = PowerTransformer(method='box-cox')
# by default meythod used is yeo-johnson

#  Box-Cox doesn't work on zero or negative values, we add a very small number (0.000001) to avoid errors.
X_train_transformed = pt.fit_transform(x_train+0.000001)
X_test_transformed = pt.transform(x_test+0.000001)

# array  mei aayega isliye dataframe mie convert kr rhe h yha
#  Box-Cox uses a special parameter called lambda (Î») to transform data.
#  pt.lambdas_ stores the lambda values for each feature (column).
pd.DataFrame({'cols':x_train.columns,'box_cox_lambdas':pt.lambdas_})
```

```python
# Applying linear regression on transformed data

lr = LinearRegression()
lr.fit(X_train_transformed,y_train)

y_pred2 = lr.predict(X_test_transformed)

r2_score(y_test,y_pred2)
```

```python
# Using cross val score

pt = PowerTransformer(method='box-cox')
X_transformed = pt.fit_transform(x+0.0000001)

lr = LinearRegression()
np.mean(cross_val_score(lr,X_transformed,y,cv=5,scoring='r2'))
```

```python
# Apply Yeo-Johnson transform

pt1 = PowerTransformer()

X_train_transformed2 = pt1.fit_transform(x_train)
X_test_transformed2 = pt1.transform(x_test)

lr = LinearRegression()
lr.fit(X_train_transformed2,y_train)

y_pred3 = lr.predict(X_test_transformed2)

print(r2_score(y_test,y_pred3))

pd.DataFrame({'cols':x_train.columns,'Yeo_Johnson_lambdas':pt1.lambdas_})
```
