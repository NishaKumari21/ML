```python
# Step 1: Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python


# Step 2: Create an Unsupervised Dataset (no labels!)
x, _ = make_moons(n_samples=300, noise=0.1, random_state=42)  # _ means we ignore labels


```

```python
# Step 3: Normalize the Data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
```

```python

# Step 4: Apply DBSCAN
db = DBSCAN(eps=0.3, min_samples=5)
clusters = db.fit_predict(X_scaled)

```

```python
# Step 5: Plot the Result
plt.figure(figsize=(8,6))
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=clusters, cmap='rainbow', s=50)
plt.title("DBSCAN Clustering on Unsupervised 'Moons' Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```
