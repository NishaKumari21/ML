```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
```

```python
# df=make_blobs()
# x=df.data
# y=df.target
# Actually, make_blobs() function tuple return karta hai — (X, y) — na ki .data aur .target jaise load_iris() mein hota hai.
```

```python
x,y=make_blobs(random_state=42,n_samples=100,n_features=2,centers=3)
```

Nahi, DBSCAN jaise unsupervised learning algorithms mein training/testing split ka concept nahi hota — kyunki:

Koi label (y) hota hi nahi to "train" karne ko kuch nahi hota.

DBSCAN sara data ek saath dekhta hai aur clusters banata hai based on density.

Isme model fit + predict ek hi step mein hota hai (not like supervised learning).

```python

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
from sklearn.cluster import DBSCAN
```

```python
db=DBSCAN(eps=0.5,min_samples=5)
```

```python
cluster=db.fit_predict(x_scaled)
```

```python
import matplotlib.pyplot as plt
# x_scaled[:, 0]: All rows, but only column 0 (i.e., first feature)
# x_scaled[:, 1] :All rows,but only second column
plt.scatter(x_scaled[:, 0], x_scaled[:, 1], c=cluster, cmap='viridis',s=50)
plt.title("DBSCAN on make_blobs Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```
